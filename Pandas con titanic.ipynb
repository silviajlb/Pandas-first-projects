Crea una columna que sea 'categoria de edad' y clasifique los pasajeros en 3 grupos segun su edad: Menor de 30, entre 30 y 55 y mayor de 55.

```python
#Solución 1
```

```python
Menor_30 = df[df['Age'] < 30]
Entre_30_55 = df[(df['Age'] >= 30) & (df['Age'] <= 55)]
Mayor_55 = df[df['Age'] > 55]
```

```python
Menor_30['categoria de edad'] = 'Menores 30'
Menor_30.head(2)
```

```python
Entre_30_55['categoria de edad'] = 'Entre 30 y 55'
Mayor_55['categoria de edad'] = 'Mayor 55'
```

```python
len(df)
```

```python
nulos = df[df['Age'].isna()]   #tambien tenemos que contar los valores nulos para que no descuadre
nulos
```

```python
df1 = pd.concat([Menor_30, Entre_30_55, Mayor_55, nulos], axis=0)
len(df1)
```

```python
def cat_edad(i):
  if i < 30:
   return  'Menores 30'
  elif i>=30 and i<=55:
   return   'Entre 30 y 55'
  elif i<55:
   return   'Mayores de 55'
```

```python
df['Age'].apply(cat_edad)
```

```python
df['Categoria edad']  = df['Age'].apply(cat_edad)
```
